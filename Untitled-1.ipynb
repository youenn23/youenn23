```python
class Carte:
    def __init__(self,valeur,couleur):
        dict_couleur = ['♥', '♣' , '♠', '♦'] 
        self.couleur = dict_couleur[couleur]
        self.valeur = valeur
    def __str__(self):
        return f'{self.valeur}{self.couleur} '

class Jeu_de_carte:
    def __init__(self,liste_de_toute_les_cartes):
        self.lst_carte = liste_de_toute_les_cartes
    def __str__(self):
        chaine = ''
        for carte in self.lst_carte:
            chaine = chaine + f'{carte.valeur}{carte.couleur} '
        return chaine
    def __len__(self):
        return len(self.lst_carte)
    
    def valeur_la_plus_haute(self):
        ordre = ['a']+ ['r','d','v']  + list(range(10,1,-1)) 
        for carte in ordre:
            if carte in [c.valeur for c in self.lst_carte]:
                return carte
        return 'a'

    def trouver_5_dans_7(self):


        
        pass

    def comparer_main_de_5(self, autre_jeu):
        '''return 'win' 'lose' 'draw' '''
        valeur_main = self.valeur_main()
        valeur_autre_main = autre_jeu.valeur_main()
        a = valeur_main[0]
        b = valeur_autre_main[0]
        if a>b:
            return 'win'
        if a<b:
            return 'lose'





    def valeur_main(self):
        assert len(self.lst_carte) == 5
        '''donne un nombre entre 0 et 9 selon la combinaison + donne les valeurs  + valeur haute (ex brelan=3points + brelan de 7 + valeur haute(dans ce cas un roi) donc return (3,7,'k')) '''
        ordre_des_carte = ['a']+ ['r','d','v']  + list(range(10,1,-1)) + ['a']
        #quinte flush royale ['a','r','d','v','10'] de meme couleur
        a = True
        couleur = self.lst_carte[0].couleur
        c = ['a','r','d','v','10']
        for k in range(5):
            if not(self.lst_carte[k].valeur in c and self.lst_carte[k].couleur == couleur):
                a = False
        if a: 
            return (9,)

        
        #quinte flush
        couleur = self.lst_carte[0].couleur
        valeur_haute = self.valeur_la_plus_haute()
        if 3 < ordre_des_carte.index(valeur_haute) < 9:
            a = True
            valeur_necessaire= valeur_haute[ordre_des_carte.index(valeur_haute) : ordre_des_carte.index(valeur_haute) +5]
            for carte in self.lst_carte:
                if carte.valeur in valeur_necessaire and carte.couleur == couleur :
                    valeur_necessaire.pop(carte.valeur)
                else:
                    a = False
        if a:
            return (8,valeur_haute)
        #carre
        dict = {}
        for carte in self.lst_carte:
            if not carte.valeur in dict:
                dict[carte.valeur] = 1
            else:
                dict[carte.valeur] += 1
        for val in dict.keys():
            if dict[val] == 4:
                return (7,val, [vale for vale in dict.keys() if vale != val][0])
        #full   
        for val in dict.keys():
            if dict[val] == 3 and [True for vale in dict.keys() if dict[vale]==2 and vale != val]:
                return (6,val, [vale for vale in dict.keys() if dict[vale]==2 and vale != val][0])
        #couleur
        a = True
        for carte in self.lst_carte:
            if not carte.couleur == couleur:
                a = False
        if a:
            return (5,self.lst_carte)
        
        
        #quinte
        valeur_haute = self.valeur_la_plus_haute()
        if 3 < ordre_des_carte.index(valeur_haute) < 9:
            a = True
            valeur_necessaire= valeur_haute[ordre_des_carte.index(valeur_haute) : ordre_des_carte.index(valeur_haute) +5]
            for carte in self.lst_carte:
                if carte.valeur in valeur_necessaire:
                    valeur_necessaire.pop(carte.valeur)
                else:
                    a = False
        if a:
            return (4,valeur_haute)
        
        #brelan
        for val in dict.keys():
            if dict[val] == 3:
                return (3,val, [cart.valeur for cart in self.lst_carte if cart.valeur != val and cart.valeur == max([carte.valeur for carte in self.lst_carte if carte.valeur!= val ])][0])
        #double pair
        for val in dict.keys():
            if dict[val] == 2 and [True for vale in dict.keys() if dict[vale]==2 and vale < val]:
                return (2,val, [vale for vale in dict.keys() if dict[vale]==2 and vale < val][0], [vale for vale in dict.keys() if dict[vale]==1][0])
        #pair
        for val in dict.keys():
            if dict[val] == 2:
                return (1,val, [cart.valeur for cart in self.lst_carte if cart.valeur != val and cart.valeur == max([carte.valeur for carte in self.lst_carte if carte.valeur!= val ])][0])
        return []

valeurs = ['a'] + list(range(2,11)) + ['v','d','r'] #as, nombre, figure
cartes = []
for couleur in range(4):
    for valeur in valeurs:
        cartes.append(Carte(valeur,couleur))
jdc = Jeu_de_carte(cartes)
```
